```python
# plot the CRPIX errors for the entire 6 days from tracking the first active region
```

```python
import matplotlib.pyplot as plt
import json
import numpy as np
import os
```

```python
fdir = '/data/slam/sinjan/arlongterm_hrt_wcs_corr/'

files = os.listdir(fdir)

DIDs = []
xerr = []
yerr = []

for f in files:
    if '2023' in f and 'CRPIX' in f:
        with open(fdir+f,'r') as of:
            d = json.load(of)
        for keys,vals in d.items():
            DIDs.append(keys)
            xerr.append(vals[0])
            yerr.append(vals[1])
```

```python
DIDs
```

```python
xerr
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(len(xerr)),xerr, label='X err')
plt.plot(range(len(yerr)),yerr, label = 'Y err')
plt.legend()
plt.ylabel('Pixel Error in CRPIX')
plt.xlabel('Dataset No.')
plt.xlim(0,35)
plt.show()
```
